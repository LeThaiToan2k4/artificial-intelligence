# CÂU 1 (2 ĐIỂM) – Python Cơ Bản

Viết chương trình Python thực hiện **quản lý danh sách sinh viên** với các yêu cầu sau:

**a) (0.5đ)** Định nghĩa class `SinhVien` có các thuộc tính: `mssv` (str), `ho_ten` (str), `nam_sinh` (int), `diem_tb` (float). Class phải có phương thức `__str__` trả về thông tin sinh viên theo định dạng:
```
[MSSV] Họ tên | Năm sinh: xxxx | Điểm TB: x.x
```

**b) (0.5đ)** Tạo danh sách ít nhất **5 sinh viên** (dữ liệu tự chọn). Sử dụng list comprehension để lọc ra danh sách sinh viên có **điểm trung bình >= 5.0** (đủ điều kiện lên lớp).

**c) (0.5đ)** Viết hàm `thong_ke(ds_sv)` nhận vào danh sách sinh viên, trả về dictionary chứa:
- `'tong_sv'`: tổng số sinh viên
- `'sv_len_lop'`: số sinh viên đủ điều kiện lên lớp (điểm TB >= 5.0)
- `'diem_cao_nhat'`: điểm trung bình cao nhất
- `'diem_trung_binh_lop'`: điểm trung bình của cả lớp

**d) (0.5đ)** In ra kết quả thống kê theo định dạng rõ ràng, đồng thời in danh sách sinh viên **từ 20 tuổi trở lên** (tính theo năm hiện tại là 2025).

````python
# CÂU 1 – Python Cơ Bản
# ==========================================

# a) Định nghĩa class SinhVien
class SinhVien:
    def __init__(self, mssv, ho_ten, nam_sinh, diem_tb):
        self.mssv = mssv
        self.ho_ten = ho_ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

    def __str__(self):
        return f"[{self.mssv:<6}] {self.ho_ten:<20} | Năm sinh: {self.nam_sinh:<6} | Điểm TB: {self.diem_tb:<5.1f}"

# b) Tạo danh sách sinh viên và lọc sinh viên đủ điều kiện lên lớp
ds_sv = [
    SinhVien('SV01', 'Nguyễn Nghĩa', 2004, 8.5),
    SinhVien('SV02', 'Lê Văn A', 2007, 4.0),
    SinhVien('SV03', 'Trần Thị B', 2005, 6.5),
    SinhVien('SV04', 'Phạm Văn C', 2006, 9.0),
    SinhVien('SV05', 'Hoàng Thị D', 2008, 3.5)
]
sv_du_dieu_kien = [sv for sv in ds_sv if sv.diem_tb >= 5.0]

# c) Hàm thống kê
def thong_ke(danh_sach):
    tong_sv = len(danh_sach)
    sv_len_lop = len([sv for sv in danh_sach if sv.diem_tb >= 5.0])
    diem_cao_nhat = max([sv.diem_tb for sv in danh_sach]) if tong_sv > 0 else 0
    diem_trung_binh_lop = sum([sv.diem_tb for sv in danh_sach]) / tong_sv if tong_sv > 0 else 0
    return {
        'tong_sv': tong_sv,
        'sv_len_lop': sv_len_lop,
        'diem_cao_nhat': diem_cao_nhat,
        'diem_trung_binh_lop': diem_trung_binh_lop
    }

# d) In kết quả thống kê và danh sách sinh viên >= 20 tuổi
kq = thong_ke(ds_sv)
print("=== KẾT QUẢ THỐNG KÊ ===")
print(f"Tổng số SV: {kq['tong_sv']}")
print(f"Số SV lên lớp: {kq['sv_len_lop']}")
print(f"Điểm cao nhất: {kq['diem_cao_nhat']}")
print(f"Điểm TB lớp: {kq['diem_trung_binh_lop']:.2f}\n")

print("=== DANH SÁCH SINH VIÊN >= 20 TUỔI (Năm 2025) ===")
for sv in ds_sv:
    if 2025 - sv.nam_sinh >= 20:
        print(sv)
````

# CÂU 2 (2.5 ĐIỂM) – Thuật Toán Tìm Kiếm A*

Cho đồ thị vô hướng có trọng số như sau (chi phí di chuyển giữa các đỉnh):

```
        A
       / \
      2   4
     /     \
    B       C
   / \     / \
  3   1   2   5
 /     \ /     \
D       E       F
         \
          3
           \
            G  (Goal)
```

**Danh sách cạnh (cạnh hai chiều):**
```
A-B: 2,  A-C: 4
B-D: 3,  B-E: 1
C-E: 2,  C-F: 5
E-G: 3
```

**Giá trị heuristic h(n) (ước lượng khoảng cách đến đích G):**
```
h(A)=7, h(B)=5, h(C)=6, h(D)=9, h(E)=3, h(F)=8, h(G)=0
```

**a) (1.0đ)** Cài đặt thuật toán A* để tìm đường đi ngắn nhất từ đỉnh **A** đến đỉnh **G**. Sử dụng danh sách `OPEN` và `CLOSE`. Với mỗi bước, in ra trạng thái của `OPEN` (node, f, g, h).

**b) (1.0đ)** In ra đường đi tối ưu tìm được và tổng chi phí thực tế.

**c) (0.5đ)** So sánh thuật toán A* với BFS (tìm kiếm theo chiều rộng): A* có ưu điểm gì? Viết câu trả lời dưới dạng comment trong cell bên dưới.

````python
# CÂU 2 – Thuật Toán Tìm Kiếm A*
# ==========================================
import heapq

# Định nghĩa đồ thị: {node: [(neighbor, cost), ...]}
graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('A', 2), ('D', 3), ('E', 1)],
    'C': [('A', 4), ('E', 2), ('F', 5)],
    'D': [('B', 3)],
    'E': [('B', 1), ('C', 2), ('G', 3)],
    'F': [('C', 5)],
    'G': [('E', 3)]
}

# Giá trị heuristic
heuristic = {
    'A': 7, 'B': 5, 'C': 6, 'D': 9,
    'E': 3, 'F': 8, 'G': 0
}

# a) Cài đặt thuật toán A*
def astar(graph, heuristic, start, goal):
    open_list = []
    # Mỗi phần tử trong open_list: (f, g, node_hiện_tại, đường_đi_đến_hiện_tại)
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))
    closed_set = set()
    
    step = 1
    while open_list:
        # Trích xuất trạng thái OPEN để in
        open_state = [(n, f, g, heuristic[n]) for f, g, n, _ in open_list]
        print(f"Bước {step} - Trạng thái OPEN (node, f, g, h): {open_state}")
        
        f, g, current, path = heapq.heappop(open_list)
        step += 1
        
        if current == goal:
            return path, g
            
        if current in closed_set:
            continue
        closed_set.add(current)
        
        for neighbor, cost in graph[current]:
            if neighbor not in closed_set:
                new_g = g + cost
                new_f = new_g + heuristic[neighbor]
                new_path = list(path)
                new_path.append(neighbor)
                heapq.heappush(open_list, (new_f, new_g, neighbor, new_path))
                
    return None, -1

# b) Gọi hàm và in kết quả
print("=" * 50)
print("THUẬT TOÁN A* – Tìm đường từ A đến G")
print("=" * 50)
path, cost = astar(graph, heuristic, 'A', 'G')

print("\n--- KẾT QUẢ ---")
if path:
    print(f"Đường đi tối ưu: {' -> '.join(path)}")
    print(f"Tổng chi phí: {cost}")
else:
    print("Không tìm thấy đường đi.")

# c) So sánh A* với BFS
# - BFS: Duyệt mù mở rộng đều đặn theo từng mức độ sâu ra mọi hướng mà không sử dụng thêm thông tin ước lượng nào. 
#        Điều này dẫn đến việc tốn rất nhiều thời gian và bộ nhớ để tìm đích trong không gian lớn.
# - A*: Là thuật toán duyệt ưu tiên có thông tin (Heuristic). Thuật toán kết hợp chi phí thực tế g(n) và ước lượng h(n) 
#       để đánh giá độ hứa hẹn của từng node (thông qua f(n) = g(n) + h(n)).
# - Ưu điểm của A* so với BFS: A* tìm được đường đi tối ưu nhanh hơn nhiều nhờ cắt bỏ/trì hoãn những nhánh đi không tốt 
#                              (nhờ hàm h(n) hướng trực tiếp về đích), giúp tiết kiệm tối đa tài nguyên tính toán và bộ nhớ.
````

# CÂU 3 (2.5 ĐIỂM) – Thuật Toán Minimax cho TicTacToe

Cài đặt trò chơi **TicTacToe 3x3** với AI sử dụng thuật toán **Minimax**.

**a) (0.5đ)** Cài đặt các hàm khởi tạo và kiểm tra trạng thái:
- `initial_state()`: trả về bảng 3x3 khởi tạo với giá trị `None`
- `player(board)`: xác định lượt đi (`'X'` hoặc `'O'`)
- `actions(board)`: trả về tập hợp các nước đi hợp lệ `(i, j)`
- `result(board, action)`: trả về bảng mới sau khi thực hiện nước đi (dùng `copy.deepcopy`)

**b) (0.5đ)** Cài đặt hàm đánh giá kết quả:
- `winner(board)`: trả về `'X'`, `'O'` hoặc `None`
- `terminal(board)`: trả về `True` nếu trò chơi kết thúc
- `utility(board)`: trả về `1` (X thắng), `-1` (O thắng), `0` (hòa)

**c) (1.0đ)** Cài đặt thuật toán Minimax:
- `max_value(state)`: tính giá trị MAX (người chơi X – muốn tối đa hóa)
- `min_value(state)`: tính giá trị MIN (AI O – muốn tối thiểu hóa)
- `minimax(board)`: trả về nước đi tối ưu cho lượt hiện tại

**d) (0.5đ)** Viết vòng lặp trò chơi đơn giản:
- Người chơi nhập tọa độ `(hàng, cột)` theo cú pháp: `1 1` (tách bởi dấu cách)
- AI tự động tính nước đi tối ưu bằng Minimax
- In bảng sau mỗi lượt; thông báo kết quả khi trò chơi kết thúc

````python
# CÂU 3 – Minimax TicTacToe
# ==========================================
import copy
import math

X = "X"
O = "O"
EMPTY = None

# a) Hàm khởi tạo và hàm hỗ trợ
def initial_state():
    """Trả về bảng 3x3 khởi tạo."""
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """Xác định lượt chơi hiện tại: X hoặc O."""
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return O if x_count > o_count else X

def actions(board):
    """Trả về tập hợp các nước đi hợp lệ (i, j)."""
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY}

def result(board, action):
    """Trả về bảng mới sau khi thực hiện action."""
    if board[action[0]][action[1]] is not EMPTY:
        raise ValueError("Invalid action")
    new_board = copy.deepcopy(board)
    new_board[action[0]][action[1]] = player(board)
    return new_board

# b) Hàm đánh giá kết quả
def winner(board):
    """Kiểm tra người chiến thắng. Trả về 'X', 'O' hoặc None."""
    for row in board:
        if row[0] == row[1] == row[2] and row[0] is not EMPTY:
            return row[0]
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] is not EMPTY:
            return board[0][col]
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] is not EMPTY:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] is not EMPTY:
        return board[0][2]
    return None

def terminal(board):
    """Trả về True nếu trò chơi kết thúc."""
    if winner(board) is not None:
        return True
    return all(cell is not EMPTY for row in board for cell in row)

def utility(board):
    """Trả về 1 (X thắng), -1 (O thắng), 0 (hòa)."""
    win = winner(board)
    if win == X:
        return 1
    elif win == O:
        return -1
    return 0

# c) Thuật toán Minimax
def max_value(state):
    """Tính giá trị tối đa cho người chơi MAX (X)."""
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, min_value(result(state, action)))
    return v

def min_value(state):
    """Tính giá trị tối thiểu cho người chơi MIN (O)."""
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, max_value(result(state, action)))
    return v

def minimax(board):
    """Trả về nước đi tối ưu (i, j) cho lượt hiện tại."""
    if terminal(board):
        return None
    current_player = player(board)
    if current_player == X:
        best_val = -math.inf
        best_action = None
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val = val
                best_action = action
        return best_action
    else:
        best_val = math.inf
        best_action = None
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val = val
                best_action = action
        return best_action

# d) Vòng lặp trò chơi
def print_board(board):
    """In bảng TicTacToe."""
    print("\n  0 1 2")
    for i, row in enumerate(board):
        row_str = f"{i} "
        for cell in row:
            row_str += (cell if cell else '.') + ' '
        print(row_str)
    print()

def play_game():
    """Vòng lặp trò chơi: Người (X) vs AI (O)."""
    board = initial_state()
    print("=" * 40)
    print("TICTACTOE – Người (X) vs AI (O)")
    print("Nhập tọa độ theo cú pháp: hàng cột (vd: 0 1)")
    print("=" * 40)
    print_board(board)

    while not terminal(board):
        current = player(board)
        if current == X:
            try:
                move = input("Lượt của bạn (X): ")
                i, j = map(int, move.split())
                if (i, j) in actions(board):
                    board = result(board, (i, j))
                else:
                    print("Nước đi không hợp lệ, hãy thử lại!")
            except:
                print("Nhập sai cú pháp! Hãy nhập hàng cột (VD: 0 1)")
        else:
            print("AI đang tính toán...")
            action = minimax(board)
            board = result(board, action)
        print_board(board)

    win = winner(board)
    if win == X:
        print("Chúc mừng! Bạn đã thắng!")
    elif win == O:
        print("Rất tiếc! AI thắng!")
    else:
        print("Trận đấu kết thúc với kết quả Hòa!")

# Bỏ comment dòng dưới để chạy trò chơi (tắt khi nộp bài)
# play_game()
````

# CÂU 4 (3 ĐIỂM) – Alpha-Beta Pruning

Cải tiến chương trình TicTacToe ở Câu 3 bằng cách thay thế Minimax thuần túy bằng **Minimax kết hợp Alpha-Beta Pruning**.

**a) (1.0đ)** Cài đặt hàm `minimax_ab(board, depth, alpha, beta, is_maximizing)` với các tham số:
- `board`: trạng thái bảng hiện tại
- `depth`: độ sâu hiện tại (dùng để ưu tiên thắng nhanh / thua muộn)
- `alpha`: cận dưới của nút MAX (`-math.inf` ban đầu)
- `beta`: cận trên của nút MIN (`+math.inf` ban đầu)
- `is_maximizing`: `True` nếu đang tính lượt MAX

Yêu cầu:
- Điểm trả về: `10 - depth` (nút MAX thắng), `-10 + depth` (nút MIN thắng), `0` (hòa)
- Cắt tỉa khi `beta <= alpha`

**b) (0.5đ)** Cài đặt hàm `find_best_move(board)` trả về nước đi tốt nhất cho AI (người chơi O – MIN) bằng cách gọi `minimax_ab`.

**c) (1.0đ)** So sánh hiệu năng giữa Minimax và Alpha-Beta Pruning:
- Thêm bộ đếm `node_count` vào cả hai hàm để đếm số lần gọi đệ quy
- Chạy cả hai từ trạng thái khởi đầu và so sánh số node được duyệt
- In kết quả so sánh

**d) (0.5đ)** Tích hợp `find_best_move` vào vòng lặp trò chơi (thay thế `minimax` ở Câu 3) và chạy thử. Nhận xét bằng comment: Alpha-Beta Pruning có tác động thế nào đến tốc độ phản hồi của AI?

````python
# CÂU 4 – Alpha-Beta Pruning
# ==========================================
# (Sử dụng lại các hàm từ Câu 3: initial_state, player, actions, result,
#  winner, terminal, print_board)
import math

# Bộ đếm node toàn cục
minimax_nodes = 0
ab_nodes = 0

# a) Minimax với Alpha-Beta Pruning
def minimax_ab(board, depth, alpha, beta, is_maximizing):
    global ab_nodes
    ab_nodes += 1

    win = winner(board)
    if win == X:
        return 10 - depth
    elif win == O:
        return -10 + depth
    elif terminal(board):
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for action in actions(board):
            eval = minimax_ab(result(board, action), depth + 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for action in actions(board):
            eval = minimax_ab(result(board, action), depth + 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return min_eval

# Minimax thuần túy (để so sánh) – CÓ đếm node
def minimax_count(board, depth, is_maximizing):
    global minimax_nodes
    minimax_nodes += 1

    win = winner(board)
    if win == X:
        return 10 - depth
    elif win == O:
        return -10 + depth
    elif terminal(board):
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for action in actions(board):
            eval = minimax_count(result(board, action), depth + 1, False)
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = math.inf
        for action in actions(board):
            eval = minimax_count(result(board, action), depth + 1, True)
            min_eval = min(min_eval, eval)
        return min_eval

# b) Tìm nước đi tốt nhất cho AI bằng Alpha-Beta
def find_best_move(board):
    best_val = math.inf
    best_action = None
    for action in actions(board):
        # Sau khi O (AI) đi, đến lượt X (người), vì vậy lượt tiếp theo is_maximizing = True
        val = minimax_ab(result(board, action), 1, -math.inf, math.inf, True)
        if val < best_val:
            best_val = val
            best_action = action
    return best_action

# c) So sánh số node giữa Minimax thuần túy và Alpha-Beta
def compare_algorithms():
    global minimax_nodes, ab_nodes
    board = initial_state()

    # Chạy Minimax thuần túy (Mô phỏng lượt của X đi trước)
    minimax_nodes = 0
    minimax_count(board, 0, True)

    # Chạy Alpha-Beta Pruning
    ab_nodes = 0
    minimax_ab(board, 0, -math.inf, math.inf, True)

    # In kết quả
    print("=" * 50)
    print("SO SÁNH MINIMAX vs ALPHA-BETA PRUNING")
    print("=" * 50)
    print(f"Minimax thuần túy  – Số node duyệt: {minimax_nodes:,}")
    print(f"Alpha-Beta Pruning – Số node duyệt: {ab_nodes:,}")
    if minimax_nodes > 0:
        reduction = (1 - ab_nodes / minimax_nodes) * 100
        print(f"Alpha-Beta cắt giảm được: {reduction:.1f}% số node")

compare_algorithms()

# d) Tích hợp Alpha-Beta vào vòng lặp trò chơi
def play_game_ab():
    board = initial_state()
    print("\n" + "=" * 40)
    print("TICTACTOE – Người (X) vs AI-AlphaBeta (O)")
    print("=" * 40)
    print_board(board)

    while not terminal(board):
        current = player(board)
        if current == X:
            try:
                move = input("Lượt của bạn (X): ")
                i, j = map(int, move.split())
                if (i, j) in actions(board):
                    board = result(board, (i, j))
                else:
                    print("Nước đi không hợp lệ!")
            except:
                print("Nhập sai cú pháp! Hãy nhập hàng cột (VD: 0 1)")
        else:
            print("AI đang tính toán bằng Alpha-Beta...")
            action = find_best_move(board)
            board = result(board, action)
        print_board(board)

    win = winner(board)
    if win == X:
        print("Chúc mừng! Bạn đã thắng!")
    elif win == O:
        print("Rất tiếc! AI thắng!")
    else:
        print("Hòa!")

    # Nhận xét:
    # Thuật toán Alpha-Beta Pruning giúp AI phản hồi gần như ngay lập tức so với Minimax thuần túy. 
    # Nhờ loại bỏ các nhánh cây tìm kiếm thừa (những trạng thái mà thuật toán biết chắc chắn là tồi tệ hơn 
    # các lựa chọn đã đánh giá), thuật toán giúp cắt giảm trên 90% lượng node cần duyệt (tính từ nước cờ đầu tiên), 
    # từ đó tăng tốc tính toán mà vẫn đảm bảo tính tối ưu của nước đi so với thuật toán gốc.

# Bỏ comment dòng dưới để chạy trò chơi (tắt khi nộp bài)
# play_game_ab()
````
