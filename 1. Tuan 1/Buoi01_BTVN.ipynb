```python
#1.	Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với
#   mỗi phần tử của mảng mới là min của 2 phần tương ứng từ 2 mảng cho trước.
#   Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì
#   các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
def tron_2_mang(a, b):
  result = []
  min_len = min(len(a), len(b))

  for i in range(min_len):
    result.append(min(a[i], b[i]))

  if len(a) > min_len:
    result.extend(b[min_len:])
  elif len(b) > min_len:
    result.extend(b[min_len:])

  return result

arr_a = [3, 9, 1, 4]
arr_b = [2, 7, 4, 3, 2, 8]
result = tron_2_mang(arr_a, arr_b)

print(f"Mảng A: {arr_a}")
print(f"Mảng A: {arr_b}")
print(f"Mảng đã trộn: {result}")
```

```python
#2.	Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
import numpy as np

arr_a = [
    [33, 2, -3, 4],
    [5, 34, 3, 8],
    [9, 2, 37, 12],
    [-1, 14, 15, 39]
]
#Cấp của ma trận
n = 4

#a.	Tính tổng các phần tử thuộc tam giác trên của đường chéo phụ
#   kể cả đường chéo phụ trong ma trận vuông a cấp n.
def sum_tamgiac_tren_phu(ma_tran, n):
  sum = 0
  for i in range(n):
    for j in range(n - i):
      sum += ma_tran[i][j]

  return sum

#b.	Chuyển các phần tử âm thành trị tuyệt đối của nó trong a.
def tri_tuyet_doi(ma_tran):
  return [[abs(pt) for pt in row] for row in ma_tran]

#c.	Thay các phần tử chẵn trong a bằng số nguyên x cho trước.
x = 36
def swap_chana_x(ma_tran, x):
  return [[x if cell % 2 == 0 else cell for cell in row] for row in ma_tran ]

#d.	Kiểm tra a có toàn chẵn không?
def kt_toan_chan(ma_tran):
  for row in ma_tran:
    for pt in row:
      if pt % 2 !=0: return False
  return True

#e.	Cho ma trận vuông a cấp n, viết hàm kiểm tra a có đối xứng không.
#   Biết rằng ma trận đối xứng là ma trận có a[i][j] = a[j][i].
def kt_doi_xung(ma_tran, n):
  for i in range(n):
    for j in range(i + 1, n):
      if ma_tran[i][j] != ma_tran[j][i]: return False
  return True

#f.	Kiểm tra ma trận vuông a cấp n có đường chéo chính tăng dần không?
def kt_dcc_tang(ma_tran, n):
  for i in range(n - 1):
    if ma_tran[i][i] >= ma_tran[i + 1][i + 1]: return False
  return True

#g.	Xuất các phần tử thuộc tam giác dưới của đường chéo phụ
#   kể cả đường chéo phụ trong ma trận vuông a cấp n.
def xuat_tgd_phu(ma_tran, n):
  result = []
  for i in range(n):
    for j in range (n - 1 - i, n):
      result.append(ma_tran[i][j])

  return result

#h.	Kiểm tra ma trận vuông a cấp n có đường chéo phụ giảm dần không?
def kt_dcp_giam(ma_tran, n):
  for i in range(n - 1):
    if ma_tran[i][n - 1 - i] <= ma_tran[i + 1][n - 2 - i]: return False
  return True

def display_matrix(ma_tran):
  for row in ma_tran:
    print(row)

print("----- Ma trận gốc -----")
display_matrix(arr_a)
print("-" * 50)

print(f"a. Tổng tam giác trên đường chéo phụ: {sum_tamgiac_tren_phu(arr_a, n)}")
print(f"b. Ma trân giá trị tuyệt đối:")
display_matrix(tri_tuyet_doi(arr_a))
print(f"c. Thay số chẵn bằng {x}:")
display_matrix(swap_chana_x(arr_a, x))
print(f"d. Ma trận toàn chẵn? {'\n-> Ma trận toàn chẵn' if kt_toan_chan(arr_a) else '\n-> Ma trận không toàn chẵn'}")
print(f"e. Ma trận đối xứng? {'\n-> Ma trận có đối xúng' if kt_doi_xung(arr_a, n) else '\n-> Ma trận không đối xứng'}")
print(f"f. Đường chéo chính tăng dần? {'\n-> Có đcc tăng dần' if kt_dcc_tang(arr_a, n) else '\n-> Không có đcc tăng dần'}")
print(f"g. Các phần tử tam giác dưới đường chéo phụ: {xuat_tgd_phu(arr_a, n)}")
print(f"h. Đường chéo phụ giảm dần? {'\n-> Có đcp giảm dần' if kt_dcp_giam(arr_a, n) else '\n-> Không có đcp giảm dần'}")
```

```python
#5.	Tìm hiểu Pandas và Matplotlib trong Python
#import thư viện
import pandas as pd
import matplotlib.pyplot as plt

#Dữ liệu
data = {
    'Tháng': ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9', 'T10', 'T11', 'T12'],
    'Doanh_Thu': [12, 10, 14, 14, 21, 23, 25, 20, 18, 13, 11, 9],
    'Hoi_Vien_Moi': [20, 18, 20, 21, 35, 37, 37, 33, 27, 25, 21, 21]
}

df = pd.DataFrame(data)

print("----- Bảng dữ liệu doanh thu phòng Gym -----")
print(df)

doanh_thu_tb = df['Doanh_Thu'].mean()
print(f"\nDoanh thu trung bình mỗi tháng: {doanh_thu_tb:.2f} Triệu VNĐ")

plt.figure(figsize=(12, 6))

plt.plot(df['Tháng'], df['Doanh_Thu'], marker='o', color='blue', label='Doanh thu (Triệu VNĐ)')

plt.bar(df['Tháng'], df['Hoi_Vien_Moi'], color='orange', alpha=0.5, label='Hội viên mới')

plt.title('Doanh thu và số hội viên mới trong 12 tháng', fontsize=14)
plt.xlabel('Tháng')
plt.ylabel('Giá trị')
plt.legend()

plt.show()
```
