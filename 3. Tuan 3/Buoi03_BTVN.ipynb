Bài 3. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.

```python
import math

class TicTacToeMinimax:
    def __init__(self, n, win_length=None):
        self.n = n
        self.board = [[' ' for _ in range(n)] for _ in range(n)]
        self.win_length = win_length if win_length else (5 if n >= 5 else n)
        self.player_X = 'X'
        self.player_O = 'O'

    def print_board(self):
        print("\n   " + "   ".join([str(i) if i < 10 else str(i) for i in range(self.n)]))
        for idx, row in enumerate(self.board):
            row_idx_str = f"{idx:2d}"
            print(f"{row_idx_str} | " + " | ".join(row) + " |")
        print("   " + "-" * (self.n * 4 + 1))

    def get_empty_cells(self):
        cells = []
        for i in range(self.n):
            for j in range(self.n):
                if self.board[i][j] == ' ':
                    cells.append((i, j))
        return cells

    def check_win(self, player):
        for i in range(self.n):
            for j in range(self.n):
                if self.board[i][j] == player:
                    if j + self.win_length <= self.n and all(self.board[i][j+k] == player for k in range(self.win_length)): return True
                    if i + self.win_length <= self.n and all(self.board[i+k][j] == player for k in range(self.win_length)): return True
                    if i + self.win_length <= self.n and j + self.win_length <= self.n and all(self.board[i+k][j+k] == player for k in range(self.win_length)): return True
                    if i - self.win_length >= -1 and j + self.win_length <= self.n and all(self.board[i-k][j+k] == player for k in range(self.win_length)): return True
        return False

    def is_game_over(self):
        return self.check_win(self.player_X) or self.check_win(self.player_O) or len(self.get_empty_cells()) == 0

    def evaluate(self):
        if self.check_win(self.player_X): return 10
        elif self.check_win(self.player_O): return -10
        else: return 0

    def minimax_pure(self, depth, is_maximizing, max_depth):
        score = self.evaluate()
        if score == 10 or score == -10: return score
        empty_cells = self.get_empty_cells()
        if len(empty_cells) == 0: return 0
        if depth >= max_depth: return 0

        if is_maximizing:
            best_score = -math.inf
            for (r, c) in empty_cells:
                self.board[r][c] = self.player_X
                best_score = max(best_score, self.minimax_pure(depth + 1, False, max_depth))
                self.board[r][c] = ' '
            return best_score
        else:
            best_score = math.inf
            for (r, c) in empty_cells:
                self.board[r][c] = self.player_O
                best_score = min(best_score, self.minimax_pure(depth + 1, True, max_depth))
                self.board[r][c] = ' '
            return best_score

    def find_best_move(self, max_depth):
        best_score = -math.inf
        best_move = None
        empty_cells = self.get_empty_cells()

        if len(empty_cells) == self.n * self.n:
             return (self.n // 2, self.n // 2)

        for (r, c) in empty_cells:
            self.board[r][c] = self.player_X
            move_score = self.minimax_pure(0, False, max_depth)
            self.board[r][c] = ' '

            if move_score > best_score:
                best_score = move_score
                best_move = (r, c)

        return best_move

    def play(self, max_depth):
        print(f"\n==============================================")
        print(f"BẮT ĐẦU CHƠI TRÊN BÀN CỜ {self.n}x{self.n} (AI Depth: {max_depth})")
        print(f"Luật: {self.win_length} ô liên tiếp là thắng.")
        print("Bạn là 'O', AI là 'X'. Bạn được đi trước!")
        print(f"==============================================")

        self.print_board()

        while not self.is_game_over():
            while True:
                try:
                    move_str = input("\nNhập nước đi của bạn (Ví dụ: '1 2' là Hàng 1, Cột 2): ")
                    r, c = map(int, move_str.split())
                    if 0 <= r < self.n and 0 <= c < self.n and (r, c) in self.get_empty_cells():
                        self.board[r][c] = self.player_O
                        break
                    else:
                        print("Tọa độ không hợp lệ hoặc ô đã có người đánh. Vui lòng nhập lại!")
                except ValueError:
                    print("Lỗi cú pháp. Vui lòng nhập hàng và cột cách nhau bởi dấu cách (VD: 0 0)")

            self.print_board()
            if self.is_game_over(): break

            # Lượt của AI (X)
            print("\nAI đang suy nghĩ...")
            best_move = self.find_best_move(max_depth)
            if best_move:
                self.board[best_move[0]][best_move[1]] = self.player_X
                print(f"AI đánh vào: {best_move[0]} {best_move[1]}")

            self.print_board()

        # Kết thúc game
        if self.check_win(self.player_X):
            print("\n>>> AI THẮNG! <<<")
        elif self.check_win(self.player_O):
            print("\n>>> BẠN THẮNG! <<<")
        else:
            print("\n>>> HÒA! <<<")

while True:
    try:
        print("\n" + "*"*40)
        print(" CHƯƠNG TRÌNH TEST MINIMAX TICTACTOE")
        print("*"*40)
        user_input = int(input("Nhập kích thước ma trận N (VD: 5, 10). Nhập 0 để THOÁT: "))

        if user_input == 0:
            print("Đã thoát chương trình. Tạm biệt!")
            break

        if user_input < 3:
            print("Kích thước quá nhỏ, vui lòng nhập N >= 3.")
            continue

        if user_input <= 3:
            depth = 9
        elif user_input < 5:
            depth = 3
        elif user_input < 10:
            depth = 2
        else:
            depth = 1

        game = TicTacToeMinimax(n=user_input)
        game.play(max_depth=depth)

    except ValueError:
        print("Lỗi: Vui lòng nhập một số nguyên!")
```

```python
import math

class TicTacToeAlphaBeta:
    def __init__(self, n, win_length=None):
        self.n = n
        self.board = [[' ' for _ in range(n)] for _ in range(n)]

        if win_length:
            self.win_length = win_length
        elif n < 5:
            self.win_length = n
        elif n == 5:
            self.win_length = 4
        else:
            self.win_length = 5

        self.player_X = 'X'
        self.player_O = 'O'

    def print_board(self):
        print("\n   " + "   ".join([str(i) if i < 10 else str(i) for i in range(self.n)]))
        for idx, row in enumerate(self.board):
            row_idx_str = f"{idx:2d}"
            print(f"{row_idx_str} | " + " | ".join(row) + " |")
        print("   " + "-" * (self.n * 4 + 1))

    def get_empty_cells(self):
        cells = []
        for i in range(self.n):
            for j in range(self.n):
                if self.board[i][j] == ' ':
                    cells.append((i, j))
        return cells

    def check_win(self, player):
        for i in range(self.n):
            for j in range(self.n):
                if self.board[i][j] == player:
                    if j + self.win_length <= self.n and all(self.board[i][j+k] == player for k in range(self.win_length)): return True
                    if i + self.win_length <= self.n and all(self.board[i+k][j] == player for k in range(self.win_length)): return True
                    if i + self.win_length <= self.n and j + self.win_length <= self.n and all(self.board[i+k][j+k] == player for k in range(self.win_length)): return True
                    if i - self.win_length >= -1 and j + self.win_length <= self.n and all(self.board[i-k][j+k] == player for k in range(self.win_length)): return True
        return False

    def is_game_over(self):
        return self.check_win(self.player_X) or self.check_win(self.player_O) or len(self.get_empty_cells()) == 0

    def evaluate(self):
        if self.check_win(self.player_X): return 10
        elif self.check_win(self.player_O): return -10
        else: return 0
    def alphabeta(self, depth, is_maximizing, alpha, beta, max_depth):
        score = self.evaluate()
        if score == 10 or score == -10: return score
        empty_cells = self.get_empty_cells()
        if len(empty_cells) == 0: return 0
        if depth >= max_depth: return 0

        if is_maximizing:
            best_score = -math.inf
            for (r, c) in empty_cells:
                self.board[r][c] = self.player_X
                eval_score = self.alphabeta(depth + 1, False, alpha, beta, max_depth)
                self.board[r][c] = ' '

                best_score = max(best_score, eval_score)
                alpha = max(alpha, eval_score)

                if beta <= alpha:
                    break
            return best_score
        else:
            best_score = math.inf
            for (r, c) in empty_cells:
                self.board[r][c] = self.player_O
                eval_score = self.alphabeta(depth + 1, True, alpha, beta, max_depth)
                self.board[r][c] = ' '

                best_score = min(best_score, eval_score)
                beta = min(beta, eval_score)

                if beta <= alpha:
                    break
            return best_score

    def find_best_move(self, max_depth):
        best_score = -math.inf
        best_move = None
        alpha = -math.inf
        beta = math.inf
        empty_cells = self.get_empty_cells()

        if len(empty_cells) == self.n * self.n:
             return (self.n // 2, self.n // 2)

        for (r, c) in empty_cells:
            self.board[r][c] = self.player_X
            move_score = self.alphabeta(0, False, alpha, beta, max_depth)
            self.board[r][c] = ' '

            if move_score > best_score:
                best_score = move_score
                best_move = (r, c)

            alpha = max(alpha, best_score)

        return best_move

    def play(self, max_depth):
        print(f"\n==============================================")
        print(f"BẮT ĐẦU CHƠI TRÊN BÀN CỜ {self.n}x{self.n} (Cắt tỉa Alpha-Beta | Depth: {max_depth})")
        print(f"Luật: {self.win_length} ô liên tiếp là thắng.")
        print("Bạn là 'O', AI là 'X'. Bạn được đi trước!")
        print(f"==============================================")

        self.print_board()

        while not self.is_game_over():
            while True:
                try:
                    move_str = input("\nNhập nước đi của bạn (Ví dụ: '1 2'): ")
                    r, c = map(int, move_str.split())
                    if 0 <= r < self.n and 0 <= c < self.n and (r, c) in self.get_empty_cells():
                        self.board[r][c] = self.player_O
                        break
                    else:
                        print("Tọa độ không hợp lệ hoặc ô đã có người đánh. Vui lòng nhập lại!")
                except ValueError:
                    print("Lỗi cú pháp. Vui lòng nhập hàng và cột cách nhau bởi dấu cách (VD: 0 0)")

            self.print_board()
            if self.is_game_over(): break

            print("\nAI đang dùng Alpha-Beta suy nghĩ...")
            best_move = self.find_best_move(max_depth)
            if best_move:
                self.board[best_move[0]][best_move[1]] = self.player_X
                print(f"AI đánh vào: {best_move[0]} {best_move[1]}")

            self.print_board()

        if self.check_win(self.player_X):
            print("\n>>> AI THẮNG! <<<")
        elif self.check_win(self.player_O):
            print("\n>>> BẠN THẮNG! <<<")
        else:
            print("\n>>> HÒA! <<<")

while True:
    try:
        print("\n" + "*"*45)
        print(" CHƯƠNG TRÌNH TEST ALPHA-BETA TICTACTOE")
        print("*"*45)
        user_input = int(input("Nhập kích thước ma trận N (VD: 5, 10). Nhập 0 để THOÁT: "))

        if user_input == 0:
            print("Đã thoát chương trình. Tạm biệt!")
            break

        if user_input < 3:
            print("Kích thước quá nhỏ, vui lòng nhập N >= 3.")
            continue

        if user_input <= 3:
            depth = 9
        elif user_input <= 5:
            depth = 3
        else:
            depth = 2

        game = TicTacToeAlphaBeta(n=user_input)
        game.play(max_depth=depth)

    except ValueError:
        print("Lỗi: Vui lòng nhập một số nguyên!")
```

```python
import chess
import math
import time


class ChessAlphaBetaAI:
    def __init__(self, depth=3):
        self.depth = depth
        self.piece_values = {
            chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 3,
            chess.ROOK: 5, chess.QUEEN: 9, chess.KING: 0
        }

    def evaluate(self, board):
        if board.is_checkmate():
            return -9999 if board.turn else 9999
        if board.is_stalemate() or board.is_insufficient_material():
            return 0

        score = 0
        for square in chess.SQUARES:
            piece = board.piece_at(square)
            if piece:
                val = self.piece_values.get(piece.piece_type, 0)
                score += val if piece.color == chess.WHITE else -val
        return score

    def alphabeta(self, board, depth, alpha, beta, is_maximizing):
        if depth == 0 or board.is_game_over():
            return self.evaluate(board)

        if is_maximizing:
            best_score = -math.inf
            for move in board.legal_moves:
                board.push(move)
                score = self.alphabeta(board, depth - 1, alpha, beta, False)
                board.pop()
                best_score = max(best_score, score)
                alpha = max(alpha, score)
                if beta <= alpha: break
            return best_score
        else:
            best_score = math.inf
            for move in board.legal_moves:
                board.push(move)
                score = self.alphabeta(board, depth - 1, alpha, beta, True)
                board.pop()
                best_score = min(best_score, score)
                beta = min(beta, score)
                if beta <= alpha: break
            return best_score

    def find_best_move(self, board):
        best_move = None
        alpha, beta = -math.inf, math.inf
        is_maximizing = board.turn == chess.WHITE

        if is_maximizing:
            best_score = -math.inf
            for move in board.legal_moves:
                board.push(move)
                score = self.alphabeta(board, self.depth - 1, alpha, beta, False)
                board.pop()
                if score > best_score:
                    best_score = score
                    best_move = move
                alpha = max(alpha, best_score)
        else:
            best_score = math.inf
            for move in board.legal_moves:
                board.push(move)
                score = self.alphabeta(board, self.depth - 1, alpha, beta, True)
                board.pop()
                if score < best_score:
                    best_score = score
                    best_move = move
                beta = min(beta, best_score)

        return best_move

    def play(self):
        board = chess.Board()
        print("\n=== BẮT ĐẦU GAME CỜ VUA (AI CẮT TỈA ALPHA-BETA) ===")
        print("Bạn cầm quân TRẮNG (Chữ hoa). AI cầm quân ĐEN (Chữ thường).")
        print("Nhập nước đi theo định dạng UCI (Ví dụ: e2e4, g1f3). Mất Vua là thua!\n")

        while not board.is_game_over():
            print(board)
            if board.turn == chess.WHITE:
                while True:
                    move_str = input("\nNhập nước đi của bạn (VD: e2e4) hoặc gõ 'quit' để thoát: ")
                    if move_str.lower() == 'quit': return
                    try:
                        move = chess.Move.from_uci(move_str)
                        if move in board.legal_moves:
                            board.push(move)
                            break
                        else:
                            print("Nước đi không hợp lệ theo luật cờ vua! Vui lòng thử lại.")
                    except:
                        print("Sai định dạng! Phải nhập theo chuẩn UCI (VD: e2e4).")
            else:
                print("\nAI đang dùng Alpha-Beta suy nghĩ...")
                start_time = time.time()
                ai_move = self.find_best_move(board)
                board.push(ai_move)
                print(f"AI quyết định đi: {ai_move} (Tốn {round(time.time() - start_time, 2)} giây)\n")
                print("-" * 30)

        print("\n=== KẾT THÚC VÁN CỜ ===")
        print("Kết quả:", board.result())

class XiangqiAlphaBetaTemplate:
    def evaluate_xiangqi_board(self, board_state):
        score = 0
        return score

    def alphabeta_xiangqi(self, board_state, depth, alpha, beta, is_max):
        pass

class GoAlphaBetaTemplate:
    def evaluate_go_board(self, board_state):
        territory_score = 0
        return territory_score

    def alphabeta_go(self, board_state, depth, alpha, beta, is_max):
        pass

if __name__ == "__main__":
    game = ChessAlphaBetaAI(depth=3)
    game.play()
```
